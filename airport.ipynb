```python
import requests
import os

if not os.path.isfile("Airport_Data.xlsx"):
    r = requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/analysing-spreadsheet-data-with-python/Airport_Data.xlsx")
    f = open("Airport_Data.xlsx","wb+")
    f.write(r.content)
    f.close()
```

```python
import pandas as pd

df_facilities = pd.read_excel("Airport_Data.xlsx", sheet_name="Facilities")
df_facilities
```

What are the types for aviation facilities in Alaska?

```python
df_facilities["Type"].unique()
```

Provide a table containing the information for all Seaplane Bases in Alaska.

```python
df_facilities[df_facilities["Type"] == "SEAPLANE BASE"]
```

What are the 5 highest aviation facilities in Alaska?

```python
df_facilities.sort_values(by = "ARPElevation", ascending = False).head(n=5)
```

What is the average elevation of aviation facilities?

```python
df_facilities["ARPElevation"].mean()
```

What are the highest and lowest elevation values of the aviation facilities?

```python
print("Highest elevation:", df_facilities["ARPElevation"].max())
print("Lowest elevation:", df_facilities["ARPElevation"].min())
```

Store the information for all Seaplane Bases in a file called seaplane.xlsx.

```python
df_seaplane = df_facilities[df_facilities["Type"] == "SEAPLANE BASE"]
df_seaplane.to_excel("seaplane.xlsx")
```
